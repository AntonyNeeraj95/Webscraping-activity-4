```python
# Importing Libraries
import selenium
import pandas as pd
import time as T

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import NoSuchElementException
```

```python
#first connecting to the webdriver
driver=webdriver.Chrome("chromedriver.exe")
url=('https://www.wikipedia.org/')
driver.get(url)
T.sleep(3)
#searching the element in the webpage
search_box=driver.find_element_by_id("searchInput")
search_box.send_keys("most viewed videos on YouTube")
search_btn=driver.find_element_by_xpath("//button[@type='submit']")
search_btn.click()
```

```python
Name=[]
Rank=[]
Artist=[]
Upload_date=[]
Views=[]
#extracting the table
table=driver.find_element_by_xpath("//div[@class='mw-parser-output']/table[3]")
#creating variables for xpath location
before_XPath = "//*[@class='mw-parser-output']/table[3]/tbody/tr["            #xpath to access every rows
aftertd_XPath_1 = "]/td[1]"                                                   #xpath to access column1                                                  
aftertd_XPath_2 = "]/td[2]"                                                   #xpath to access column2
aftertd_XPath_3 = "]/td[3]"                                                   #xpath to access column3
aftertd_XPath_4 = "]/td[4]"                                                   #xpath to access column4
aftertd_XPath_5 = "]/td[5]"                                                   #xpath to access column5
rows = len(driver.find_elements_by_xpath("//*[@class='mw-parser-output']/table[3]/tbody/tr"))
columns = len(driver.find_elements_by_xpath("//*[@class='mw-parser-output']/table[3]/tbody/tr[1]/td"))
print(rows," ", columns)
```

```python
#extracting the details for rank
for row in range (1,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_1
    column_1=driver.find_element_by_xpath(FinalXPath).text
    Rank.append(column_1)

#extracting the details for name
for row in range (1,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_2
    column_2=driver.find_element_by_xpath(FinalXPath).text
    Name.append(column_2.split('"')[1])

#extracting the details for artists
for row in range (1,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_3
    column_3=driver.find_element_by_xpath(FinalXPath).text
    Artist.append(column_3)

#extracting the details for views
for row in range (1,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_4
    column_4=driver.find_element_by_xpath(FinalXPath).text
    Views.append(column_4)

#extracting the details for upload date
for row in range (1,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_5
    column_5=driver.find_element_by_xpath(FinalXPath).text
    Upload_date.append(column_5)
```

```python
driver.close()
#creating a dataframe
most_viewed=pd.DataFrame()
most_viewed['Rank']=Rank
most_viewed['Name']=Name
most_viewed['Artist']=Artist
most_viewed['Views']=Views
most_viewed['Upload_date']=Upload_date
most_viewed
```

```python
Name=[]
Designation=[]
Company=[]
Skills_required=[]
Location=[]

#extracting name
name=driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
for i in name:
    Name.append(i.text)

#extracting designation
designation=driver.find_elements_by_xpath("//span[@class='ellipsis clr']")
for i in designation:
    Designation.append(i.text)

#extracting company name
company=driver.find_elements_by_xpath("//a[@class='ellipsis']")
for i in company:
    Company.append(i.text)
Company=Company[1:101:2]

#extracting skills hired
skills=driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
for i in skills:
    Skills_required.append(i.text)

#extracting location
Xpath_1="//div[@id='tabP-1']/div["
Xpath_2="]/div["
Xpath_3="]/div[1]/div[1]/p[1]/span[2]/small"
boxes=len(driver.find_elements_by_xpath("//p[@class='highlightable']"))
for i in range(1,(int(boxes/2)+1)):
    for j in range(1,3):
        try:
            location=driver.find_element_by_xpath(Xpath_1+str(i)+Xpath_2+str(j)+Xpath_3).text
            Location.append(location)
        except NoSuchElementException as e:
            Location.append("-")
```

### 10. Details of Datasets from UCI machine learning repositories.

```python
#first connecting to the webdriver
driver=webdriver.Chrome("chromedriver.exe")
url=('https://archive.ics.uci.edu/')
driver.get(url)
T.sleep(2)
page=driver.find_element_by_xpath("//table[@bgcolor='#003366']/tbody/tr/td[2]/span[2]")
page.click()
```

```python
Name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]

#creating variables for xpath location
before_XPath = "//table[@border='1']/tbody/tr["
aftertd_XPath_1 = "]/td[1]"
aftertd_XPath_2 = "]/td[2]"
aftertd_XPath_3 = "]/td[3]"
aftertd_XPath_4 = "]/td[4]"
aftertd_XPath_5 = "]/td[5]"
aftertd_XPath_6 = "]/td[6]"
aftertd_XPath_7 = "]/td[7]"

rows=len(driver.find_elements_by_xpath("//table[@border='1']/tbody/tr"))

#extracting dataset name
for row in range (2,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_1
    column_1=driver.find_element_by_xpath(FinalXPath).text
    if column_1 is not None:
        Name.append(column_1)
    else:
        Name.append("-")

#extracting data type
for row in range (2,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_2
    column_2=driver.find_element_by_xpath(FinalXPath).text
    if column_2 is not None:
        Data_type.append(column_2)
    else:
        Data_type.append("-")

#extracting task
for row in range (2,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_3
    column_3=driver.find_element_by_xpath(FinalXPath).text
    if column_3 is not None:
        Task.append(column_3)
    else:
        Task.append("-")

#extracting Attribute type
for row in range (2,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_4
    column_4=driver.find_element_by_xpath(FinalXPath).text
    if column_4 is not None:
        Attribute_type.append(column_4)
    else:
        Attribute_type.append("-")

#extracting number of instances
for row in range (2,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_5
    column_5=driver.find_element_by_xpath(FinalXPath).text
    if column_5 is not None:
        No_of_instances.append(column_5)
    else:
        No_of_instances.append("-")

#extracting number of attributes
for row in range (2,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_6
    column_6=driver.find_element_by_xpath(FinalXPath).text
    if column_6 is not None:
        No_of_attribute.append(column_6)
    else:
        No_of_attribute.append("-")

#extracting year
for row in range (2,(rows+1)):
    FinalXPath = before_XPath + str(row) + aftertd_XPath_7
    column_7=driver.find_element_by_xpath(FinalXPath).text
    if column_7 is not None:
        Year.append(column_7)
    else:
        Year.append("-")
```

```python
driver.close()
#creating a dataframe
Datasets=pd.DataFrame()
Datasets['Dataset name']=Name
Datasets['Data type']=Data_type
Datasets['Task']=Task
Datasets['Attribute type']=Attribute_type
Datasets['No of instances']=No_of_instances
Datasets['No of attribute']=No_of_attribute
Datasets['Year']=No_of_attribute
Datasets
```
